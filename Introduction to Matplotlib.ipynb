   # Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
       xlabel = "x-axies",
       ylabel = "y-axies")

# 5. Save & show
fig.savefig("figure/sample-plot.png")
```

```python
car = car_sales.copy()
car
```

```python
car["Price"] = car["Price"].str.replace("(\.00)$","") #remove the last two zeros
car["Price"] = car["Price"].str.replace("[\D]","") #remove all non-digit character
```

```python
car
```

```python
 car["Sale Date"] = pd.date_range("1/1/2020",periods=len(car))
```

```python
car
```

```python
car["Price"] = car["Price"].astype(int)
```

```python
car["Total"] = car["Price"].cumsum()
```

```python
car
```

```python
car.plot(x="Sale Date",y="Total");
```

```python
car = car.sort_values("Odometer (KM)")
```

```python
car.plot(x="Odometer (KM)",y="Price",kind="bar");
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
len(heart_disease)
```

```python
heart_disease.head()
```

```python
heart_disease["age"].plot.hist(bins=25);
```

```python
heart_disease.plot.hist()
```

```python
fig, ax = plt.subplots(nrows=7,ncols=2)
ax[0,0] =  heart_disease["age"].plot.hist()
```

```python
over_50 = heart_disease[heart_disease["age"] >= 50]
len(over_50)
```

```python
over_50.head()
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True )

scatter = ax0.scatter(x=over_50["age"],
           y=over_50["thalach"],
           c=over_50["target"],
            cmap="winter")
ax0.set(title="Heart disease and cholesterol levels",
        ylabel="thalach")
ax0.legend(*scatter.legend_elements(),title="Target")
ax0.axhline(y=over_50["thalach"].mean(), linestyle="--")

scatter1 = ax1.scatter(x=over_50["age"],
           y=over_50["thalach"],
           c=over_50["target"],
                      cmap="winter")
ax1.set(title="Heart disease and cholesterol levels",
       xlabel="Age", ylabel="thalach")
ax1.legend(*scatter.legend_elements(),title="Target")
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--")

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
fig.savefig("figure/hearth-disease-analysis")
```
